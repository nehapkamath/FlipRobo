# 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Importing Libraries
import pandas as pd

#import library to query webpage 
import urllib.request as urllib2 
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time

#import exceptions
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python

#specifying page of interest
wiki = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
#save the HTML of the site within the page variable
page = urllib2.urlopen(wiki)

#parse data from "page" and save to new variable "soup"
soup = BeautifulSoup(page)


#pinpointing the location of the table and its contents
first_table = soup.find('table', class_ = 'wikitable sortable')
#creating lists for each column in table.
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]


#utilizing HTML tags for rows <tr> and elements <td> to iterate through each row of data and append data elements to their appropriate lists:
for row in first_table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells)==6: #to extract only table body not heading
        A.append(cells[0].find(text=True))
        B.append(cells[1].find())
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
        

#convert list to dataframe
df=pd.DataFrame(A,columns=['Rank'])
df['Name']=B
df['Artist']=C
df['Views']=D
df['UploadDate']=E
df['Note']=F

#Source: https://towardsdatascience.com/web-scraping-the-list-of-greatest-video-games-from-wikipedia-pt-1-230c64d93e8c
```
